```python
#去重
df_db.drop_duplicates()
df_db.rename(columns={"channel_bookingnumber": "Supplier Order ID"}, inplace=True)
```

```python
df_db['Supplier Order ID'] = df_db['Supplier Order ID'].apply(pd.to_numeric)
df_clean['Supplier Order ID'] = df_clean['Supplier Order ID'].apply(pd.to_numeric)
```

```python
result = pd.merge(df_db, df_clean, how='left', on=['Supplier Order ID'])
goal = result.drop(result.columns[[11,13,14]], axis=1)
```

```python
dt = goal['channel_createdate'].astype(str)
goal['year'] = dt.apply(lambda x:x.split('-')[0])
goal['month'] = dt.apply(lambda x:x.split('-')[1])
goal['week'] = pd.to_datetime(goal['channel_createdate']).dt.week
```

```python
goal['真实ttv'] = goal['Adjusted Cost']*goal['channel_exchange']
goal['原始ttv'] = goal['Initial Cost']*goal['channel_exchange']
goal['真实GP'] = goal['Adjusted Cost']*goal['channel_exchange']-goal['supplier_postbookprice']*goal['supplier_exchange']
goal['原始GP'] = goal['Initial Cost']*goal['channel_exchange']-goal['supplier_postbookprice']*goal['supplier_exchange']
goal['GP'] = goal['channel_price']*goal['channel_exchange']-goal['supplier_postbookprice']*goal['supplier_exchange']
```

```python
goal
```

```python
goal_grouped = goal.groupby(['clientid','year','month','week']).agg({'Supplier Order ID':'count','channel_pricecny':'sum','GP':'sum','原始ttv':'sum','原始GP':'sum','真实ttv':['count','sum'],'真实GP':'sum','supplier_postbookprice':'sum'}).reset_index()
goal_grouped.columns = goal_grouped.columns.map('_'.join)
goal_grouped = goal_grouped.rename(columns={'clientid_':'Client ID','year_':'Year','month_':'Month','week_':'week','Supplier Order ID_count':'Total订单数量','channel_pricecny_sum':'TotalTTV','GP_sum':'Total利润','真实ttv_count':'MRM订单数量','原始ttv_sum':'MRM原始TTV','原始GP_sum':'MRM原始GP','真实ttv_sum':'MRM真实TTV','真实GP_sum':'MRM真实GP','supplier_postbookprice_sum':'Supplier成本'})
goal_grouped
```

```python
goal_grouped['Total GPrate'] = goal_grouped['Total利润']/goal_grouped['TotalTTV']
goal_grouped['MRM原始GPrate'] = goal_grouped['MRM原始GP']/goal_grouped['MRM原始TTV']
goal_grouped['MRM真实GPrate'] = goal_grouped['MRM真实GP']/goal_grouped['MRM真实TTV']
goal_grouped['订单占比'] = goal_grouped['MRM订单数量']/goal_grouped['Total订单数量']
goal_grouped['TTV占比'] = goal_grouped['MRM真实TTV']/goal_grouped['TotalTTV']
goal_grouped['利润占比'] = goal_grouped['MRM真实GP']/goal_grouped['Total利润']
goal_grouped.fillna(0, inplace=True)
```
